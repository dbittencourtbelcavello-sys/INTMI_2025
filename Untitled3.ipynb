```python
total = 0
cont = 0

while True:
    idade = int(input("Digite a idade (negativa para sair): "))
    if idade < 0:
        break
    total += idade
    cont += 1

if cont > 0:
    media = total / cont
    print("Média das idades:", media)
else:
    print("Nenhuma idade válida foi digitada.")
```

```python
positivos = 0
negativos = 0
zeros = 0

for _ in range(10):
    num = float(input("Digite um número: "))
    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        zeros += 1

print("Positivos:", positivos)
print("Negativos:", negativos)
print("Zeros:", zeros)
```
