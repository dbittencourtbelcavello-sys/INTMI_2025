```python
numeros = [int(input("Digite um número inteiro: ")) for _ in range(10)]
def primo(n):
    if n < 2: return False
    return all(n % i for i in range(2, int(n**0.5) + 1))

primos = [n for n in numeros if primo(n)]
print("Números primos:", primos)
```

```python
m = [[float(input(f"Valor [{i}][{j}]: ")) for j in range(3)] for i in range(3)]
det = (
    m[0][0]*(m[1][1]*m[2][2] - m[1][2]*m[2][1])
    - m[0][1]*(m[1][0]*m[2][2] - m[1][2]*m[2][0])
    + m[0][2]*(m[1][0]*m[2][1] - m[1][1]*m[2][0])
)
print("Determinante:", det)3
```

```python
A = [[int(input(f"Matriz A [{i}][{j}]: ")) for j in range(3)] for i in range(3)]
B = [[int(input(f"Matriz B [{i}][{j}]: ")) for j in range(3)] for i in range(3)]
soma = [[A[i][j] + B[i][j] for j in range(3)] for i in range(3)]
diff = [[A[i][j] - B[i][j] for j in range(3)] for i in range(3)]
print("Matriz soma:", soma)
print("Matriz diferença:", diff)
```

```python
import string
frase = input("Digite uma frase: ").lower()
for p in string.punctuation:
    frase = frase.replace(p, "")
palavras = frase.split()
ocorrencias = {p: palavras.count(p) for p in set(palavras)}
print("Ocorrências:", ocorrencias)
```
